```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# data from https://www.kaggle.com/nehalbirla/vehicle-dataset-from-cardekho?select=Car+details+v3.csv
df_cars = pd.read_csv('car data.csv')

```

```python
x = df_cars["Present_Price"].values
y = df_cars["Selling_Price"].values

```

```python
# column vector
X=x.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=100)
```

```python

%matplotlib inline
plt.scatter(x_train,y_train,color='red')
plt.xlabel('Sticker Price')
plt.ylabel('Sold Price')
plt.title('Car Sale Data')
plt.show()
```

```python

lm = LinearRegression()
lm.fit(x_train,y_train)
y_predict = lm.predict(x_test)
print(f"Train accuracy {round(lm.score(x_train,y_train)*100,2)} %")
print(f"Test accuracy {round(lm.score(x_test,y_test)*100,2)} %")
```

```python

plt.scatter(x_train,y_train,color='red')
plt.plot(x_test,y_predict)
plt.xlabel("Sticker Price")
plt.ylabel("Sold Price")
plt.title("Car Sale Data")
plt.plot
```

```python
# Conclusions: There is a linear corelation between the sticker price and sale price. The sale prace is always below the sticker price. Most sales are close to 50% of the sticker price.
```
